**5. Obtener todos los datos de la tabla Person.Person y agregar una columna con la siguiente lgica: Si Email Promotion es 0 entonces escribir NO, caso contrario escribir SI.**



```
USE AdventureWorks

SELECT 
	BusinessEntityID, PersonType, NameStyle, Title, FirstName, MiddleName, LastName, Suffix, EmailPromotion, AdditionalContactInfo, Demographics, rowguid, ModifiedDate,
	CASE
		WHEN EmailPromotion = 0 THEN 'No'
		ELSE 'Si'
	END EmailLogic
FROM Person.Person
```
